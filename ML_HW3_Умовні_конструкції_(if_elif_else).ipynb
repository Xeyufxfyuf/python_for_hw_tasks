
**1. Основи роботи з умовними конструкціями (if-elif-else)**

Створіть змінну `cooking_time = 5` (час варіння яєць у хвилинах)

Напишіть програму, яка виводить (кожен пункт оцінюється в 0.5 бала):
- "Яйце рідке" якщо cooking_time < 4
- "Яйце всмятку" якщо 4 <= cooking_time < 7
- "Яйце вкруту" якщо 7 <= cooking_time < 10
- "Яйце переварене" якщо cooking_time >= 10


```python
cooking_time = 5
```

```python
def cooking_eggs(cooking_time):
    if cooking_time < 4:
        print('Яйце рідке')
    elif 4 <= cooking_time < 7:
        print('Яйце всмятку')
    elif 7 <= cooking_time < 10:
        print('Яйце вкруту')
    else:
        print('Яйце переварене')

cooking_eggs(cooking_time)        
```

**2. Перевірка оцінки**

Створіть змінну `grade = 72`

Напишіть програму, яка виводить (кожен пункт оцінюється в 0.5 бала):
- "Відмінно" якщо grade >= 90
- "Добре" якщо 75 <= grade < 90
- "Задовільно" якщо 60 <= grade < 75
- "Незадовільно" якщо grade < 60


```python
grade = 72
```

```python
def show_grade(grade):
    if grade >= 90:
        print('Відмінно')
    elif 75 <= grade < 90:
        print('Добре')
    elif 60 <= grade < 75:
        print('Задовільно')
    else:
        print('Незадовільно')

show_grade(grade)
        
```

**3. Перевірка на наявність**

Створіть список продуктів в холодильнику зі значеннями ["молоко", "яйця", "масло", "сир"] і запишіть його в змінну.


Напишіть код, який перевіряє (кожен пункт оцінюється в 0.5 бала):
- Чи є "яйця" в списку (виведіть "Яйця є в холодильнику" або "Яєць немає")
- Чи є "хліб" в списку


```python
def find_product(product):
    products = ["молоко", "яйця", "масло", "сир"]
    if product in products:
        print(product.capitalize() + ' є в холодильнику')
    else:
        print(product.capitalize() + ' немає')

find_product('яйця')
find_product('хліб')              
```
